```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:\Python\Data\Video_Games_Sales_as_at_22_Dec_2016a.csv')
print (df)
```

```python
plt.plot(df['Global_Sales'])
plt.show()
```

```python
#Extremely basic graph. Spanning several decades and covering many different gaming companies this graph shows:
#That very few games break the $10MM in sales in a single year. In fact during a single year, the majority of games will
#by far make less than $10MM in sales
```

```python
plt.figure(figsize=(10,15))
plt.subplot(2, 2, 1)
plt.plot(df['NA_Sales'])
plt.ylabel("Sales in Million")
plt.xlabel("Individual Game")
plt.title("Sales of Individual Games in North America")

plt.subplot(2, 2, 2)
plt.plot(df['EU_Sales'])
plt.ylabel("Sales in Million")
plt.xlabel("Individual Game")
plt.title("Sales of Individual Games in Europe")

plt.subplot(2, 2, 3)
plt.plot(df['JP_Sales'])
plt.ylabel("Sales in Million")
plt.xlabel("Individual Game")
plt.title("Sales of Individual Games in Japan")

plt.subplot(2, 2, 4)
plt.plot(df['Other_Sales'])
plt.ylabel("Sales in Million")
plt.xlabel("Individual Game")
plt.title("Sales of Individual Games in Other Countries")

plt.show()
```

```python
#Telescoping into each zone regionally we can see that North America is the country where perhaps there is greatest opportunity
#for large sales profits. Europe being the second best market, followed by Japan and then "other" countries.
```

```python
df1 = df[['Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
print(df1)
```

```python
Platform_Sales = df1.groupby('Platform', as_index=False).sum()
print(Platform_Sales)
```

```python
plt.plot(Platform_Sales['Global_Sales'])
plt.ylabel('Money in Millions')
plt.xlabel('Corresponding number which indicates company on the list')
plt.title("Global Sales Trends in the Gaming Industry")
plt.show()
```

```python
#Global Sales by PLATFORM. Looks like the PS2 mopped the floor with everyone else (2 out of 4 largest peaks on the graph above
#are Playstation products)
```

```python
plt.scatter(x = Platform_Sales['NA_Sales'], y = Platform_Sales['EU_Sales'])
plt.ylabel("Sales in Europe")
plt.xlabel("Sales in NA")
plt.title("Video Games Sales in the West")
plt.show()
```

```python
#Looking at sales by platform on a scatter-chart to see that there are a few games that do extremely well in both markets. 
# It would be very interesting to see Sales in relation to country population. 
```
